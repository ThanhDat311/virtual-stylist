```python
import os, json, random, time, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import torch
import torch.nn.functional as F
import faiss
import kagglehub
from PIL import Image, UnidentifiedImageError
from transformers import CLIPModel, CLIPProcessor
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from collections import Counter
from tqdm.notebook import tqdm
warnings.filterwarnings('ignore')

# ── GPU Check ────────────────────────────────────────────────────────────────
device = 'cuda' if torch.cuda.is_available() else 'cpu'
if device == 'cuda':
    print(f'✅ GPU: {torch.cuda.get_device_name(0)}')
    print(f'   VRAM: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB')
else:
    print('⚠️  GPU không tìm thấy — hãy vào Runtime → Change runtime type → T4 GPU')

# ── Output directories ───────────────────────────────────────────────────────
DATA_DIR      = '/content/data'
PROCESSED_DIR = '/content/data/processed'
OUTPUTS_DIR   = '/content/outputs'
os.makedirs(PROCESSED_DIR, exist_ok=True)
os.makedirs(OUTPUTS_DIR,   exist_ok=True)
print(f'📁 Output dirs: {DATA_DIR}, {OUTPUTS_DIR}')
```

## 📦 Bước 2 — Tải dataset từ Kaggle & lọc + resize ảnh

```python
# ── Cấu hình ─────────────────────────────────────────────────────────────────
NUM_IMAGES  = 5000   # Số ảnh tối đa (tăng lên nếu muốn toàn bộ dataset)
IMG_SIZE    = 224    # Chuẩn kích thước 224×224
BATCH_SIZE  = 64     # Batch size cho CLIP (T4 GPU xử lý tốt 64)

print('📥 Đang tải Fashion Product Images Dataset từ Kaggle...')
t0 = time.time()
dataset_path = kagglehub.dataset_download('paramaggarwal/fashion-product-images-dataset')
print(f'   Dataset path: {dataset_path}  ({time.time()-t0:.1f}s)')

IMAGES_DIR    = os.path.join(dataset_path, 'fashion-dataset/images')
STYLES_CSV    = os.path.join(dataset_path, 'fashion-dataset/styles.csv')

# ── Đọc CSV metadata ─────────────────────────────────────────────────────────
df_full = pd.read_csv(STYLES_CSV, on_bad_lines='skip')
print(f'   CSV loaded: {len(df_full)} rows | Columns: {list(df_full.columns)}')

# ── Kiểm tra file ảnh tồn tại ────────────────────────────────────────────────
print('🔍 Đang kiểm tra ảnh hợp lệ...')
df_full['image_path'] = df_full['id'].apply(
    lambda i: os.path.join(IMAGES_DIR, f'{i}.jpg')
)
df_full['file_exists'] = df_full['image_path'].apply(os.path.exists)
df_valid = df_full[df_full['file_exists']].copy()
print(f'   Hợp lệ: {len(df_valid)} / {len(df_full)} ảnh')

# ── Giới hạn số ảnh ──────────────────────────────────────────────────────────
df = df_valid.head(NUM_IMAGES).reset_index(drop=True)
print(f'\n✅ Sẽ xử lý {len(df)} ảnh')
print(f"   Thể loại: {df['masterCategory'].value_counts().to_dict()}")
display(df[['id','productDisplayName','masterCategory','subCategory','baseColour']].head(5))
```

```python
# ── Resize & lọc ảnh hỏng → data/processed/ ──────────────────────────────────
valid_rows = []
skipped    = 0

print(f'🖼️  Đang resize {len(df)} ảnh → {IMG_SIZE}×{IMG_SIZE} px...')

for _, row in tqdm(df.iterrows(), total=len(df), desc='Resize'):
    dst = os.path.join(PROCESSED_DIR, f"{row['id']}.jpg")
    if os.path.exists(dst):
        valid_rows.append({**row, 'processed_path': dst})
        continue
    try:
        with Image.open(row['image_path']) as img:
            if img.size[0] < 32 or img.size[1] < 32:  # bỏ ảnh quá nhỏ
                skipped += 1
                continue
            img.convert('RGB').resize(
                (IMG_SIZE, IMG_SIZE), Image.LANCZOS
            ).save(dst, 'JPEG', quality=90)
        valid_rows.append({**row, 'processed_path': dst})
    except (UnidentifiedImageError, OSError):
        skipped += 1

df_proc = pd.DataFrame(valid_rows).reset_index(drop=True)
print(f'\n✅ Resize xong!')
print(f'   Hợp lệ  : {len(df_proc)}')
print(f'   Bỏ qua  : {skipped} (ảnh hỏng / quá nhỏ)')
```

## 🤖 Bước 3 — Load CLIP & Extract Feature Embeddings

```python
MODEL_ID = 'openai/clip-vit-base-patch32'
print(f'⬇️  Loading {MODEL_ID} on {device}...')
t0 = time.time()
clip_model     = CLIPModel.from_pretrained(MODEL_ID).to(device)
clip_processor = CLIPProcessor.from_pretrained(MODEL_ID)
clip_model.eval()
print(f'✅ CLIP loaded in {time.time()-t0:.1f}s | dim=512')
```

```python
def extract_features_batch(image_paths, batch_size=BATCH_SIZE):
    """Extract L2-normalised 512-dim CLIP embeddings for a list of image paths."""
    all_feats = []
    total     = len(image_paths)

    for i in tqdm(range(0, total, batch_size), desc='CLIP Extraction'):
        batch_paths  = image_paths[i : i + batch_size]
        batch_images = []
        for p in batch_paths:
            try:
                batch_images.append(Image.open(p).convert('RGB'))
            except Exception:
                pass
        if not batch_images:
            continue

        inputs = clip_processor(
            images=batch_images, return_tensors='pt', padding=True
        ).to(device)

        with torch.no_grad():
            feats = clip_model.get_image_features(**inputs)
            # Robustness: handle older transformers versions
            if not isinstance(feats, torch.Tensor):
                feats = getattr(feats, 'image_embeds',
                        getattr(feats, 'pooler_output', feats[0]))
            feats = F.normalize(feats, p=2, dim=-1)  # L2 normalize → cosine

        all_feats.append(feats.cpu().numpy())

    return np.concatenate(all_feats, axis=0).astype(np.float32)

# ── Run extraction ────────────────────────────────────────────────────────────
image_paths_list = df_proc['processed_path'].tolist()
t0 = time.time()
embeddings = extract_features_batch(image_paths_list)
elapsed    = time.time() - t0

print(f'\n✅ Extraction xong!')
print(f'   Shape   : {embeddings.shape}   (N × 512)')
print(f'   Thời gian: {elapsed:.1f}s  ({elapsed/len(embeddings)*1000:.1f} ms/ảnh)')
print(f'   Norm check (should be ~1.0): {np.linalg.norm(embeddings[0]):.4f}')
```

## 🗃️ Bước 4 — Build FAISS Index & Xuất Artifacts

```python
# ── Build FAISS IndexFlatIP (Inner Product = Cosine với L2-norm data) ─────────
dim   = embeddings.shape[1]
index = faiss.IndexFlatIP(dim)
index.add(embeddings)
print(f'✅ FAISS index: {index.ntotal} vectors, dim={dim}')

# ── Build metadata dictionary ─────────────────────────────────────────────────
metadata = {}
for faiss_idx, (_, row) in enumerate(df_proc.iterrows()):
    metadata[str(faiss_idx)] = {
        'id'               : str(row.get('id', '')),
        'name'             : str(row.get('productDisplayName', '')),
        'category'         : str(row.get('masterCategory', '')),
        'sub_category'     : str(row.get('subCategory', '')),
        'article_type'     : str(row.get('articleType', '')),
        'color'            : str(row.get('baseColour', '')),
        'gender'           : str(row.get('gender', '')),
        'season'           : str(row.get('season', '')),
        'image_path'       : row.get('image_path', ''),
        'local_image_path' : row.get('processed_path', ''),
        'faiss_index'      : faiss_idx,
    }

# ── Save artifacts ────────────────────────────────────────────────────────────
FAISS_PATH      = os.path.join(DATA_DIR, 'vector_index.bin')
METADATA_PATH   = os.path.join(DATA_DIR, 'metadata.json')
EMBEDDINGS_PATH = os.path.join(DATA_DIR, 'embeddings.npy')
IMG_PATHS_PATH  = os.path.join(DATA_DIR, 'image_paths.npy')

faiss.write_index(index, FAISS_PATH)
with open(METADATA_PATH, 'w', encoding='utf-8') as f:
    json.dump(metadata, f, ensure_ascii=False)
np.save(EMBEDDINGS_PATH, embeddings)
np.save(IMG_PATHS_PATH, np.array(image_paths_list))

print(f'\n📁 Đã lưu tất cả artifacts vào {DATA_DIR}:')
for p in [FAISS_PATH, METADATA_PATH, EMBEDDINGS_PATH, IMG_PATHS_PATH]:
    size_mb = os.path.getsize(p) / 1e6
    print(f'   {os.path.basename(p):25s}  {size_mb:.2f} MB')
```

## 📊 Bước 5 — T-SNE & PCA Visualization (Milestone Check)

```python
# ── PCA (nhanh, chạy trên toàn bộ data) ──────────────────────────────────────
print('📉 Đang chạy PCA...')
pca    = PCA(n_components=2, random_state=42)
coords = pca.fit_transform(embeddings)
var    = pca.explained_variance_ratio_
print(f'   PC1: {var[0]*100:.1f}%  PC2: {var[1]*100:.1f}%  (total: {sum(var)*100:.1f}%)')

categories = [metadata[str(i)]['category'] for i in range(len(metadata))]

PALETTE = {
    'Apparel'       : '#4E79A7',
    'Footwear'      : '#F28E2B',
    'Accessories'   : '#59A14F',
    'Personal Care' : '#E15759',
    'Free Items'    : '#B07AA1',
    'Sporting Goods': '#76B7B2',
    'Home'          : '#EDC948',
}
default_color = '#9C755F'

fig, ax = plt.subplots(figsize=(13, 9), facecolor='#1a1a2e')
ax.set_facecolor('#16213e')

unique_cats = sorted(set(categories))
for cat in unique_cats:
    mask = np.array([c == cat for c in categories])
    color = PALETTE.get(cat, default_color)
    ax.scatter(coords[mask, 0], coords[mask, 1],
               c=color, s=14, alpha=0.72, linewidths=0,
               label=f'{cat} (n={mask.sum()})')

ax.set_title('Fashion Embeddings — PCA 2D\n(Milestone Check: Giai đoạn 1)',
             color='white', fontsize=15, fontweight='bold')
ax.set_xlabel(f'PC1  ({var[0]*100:.1f}% variance)', color='#aaa', fontsize=11)
ax.set_ylabel(f'PC2  ({var[1]*100:.1f}% variance)', color='#aaa', fontsize=11)
ax.tick_params(colors='#555')
for sp in ax.spines.values(): sp.set_edgecolor('#333355')
ax.legend(facecolor='#0f3460', edgecolor='#333355', labelcolor='white',
          fontsize=10, markerscale=2, loc='best', framealpha=0.3)

plt.tight_layout()
pca_out = os.path.join(OUTPUTS_DIR, 'pca_visualization.png')
plt.savefig(pca_out, dpi=150, bbox_inches='tight')
plt.show()
print(f'✅ Saved → {pca_out}')
```

```python
# ── T-SNE (chính xác hơn nhưng chậm hơn — subsample 3000 nếu cần) ────────────
MAX_TSNE = 3000
if len(embeddings) > MAX_TSNE:
    rng   = np.random.default_rng(42)
    idx   = rng.choice(len(embeddings), MAX_TSNE, replace=False)
    emb_t = embeddings[idx]
    cat_t = [categories[i] for i in idx]
    print(f'T-SNE: dùng {MAX_TSNE} mẫu (subsample)')
else:
    emb_t, cat_t = embeddings, categories

print(f'📉 Đang chạy T-SNE trên {len(emb_t)} điểm (có thể mất 2-5 phút)...')
t0 = time.time()
coords_t = TSNE(
    n_components=2, perplexity=40, n_iter=1000,
    random_state=42, verbose=0
).fit_transform(emb_t)
print(f'   Done in {time.time()-t0:.1f}s')

fig, ax = plt.subplots(figsize=(13, 9), facecolor='#1a1a2e')
ax.set_facecolor('#16213e')

for cat in sorted(set(cat_t)):
    mask  = np.array([c == cat for c in cat_t])
    color = PALETTE.get(cat, default_color)
    ax.scatter(coords_t[mask, 0], coords_t[mask, 1],
               c=color, s=14, alpha=0.72, linewidths=0,
               label=f'{cat} (n={mask.sum()})')

ax.set_title('Fashion Embeddings — T-SNE 2D\n(Milestone Check: Giai đoạn 1)',
             color='white', fontsize=15, fontweight='bold')
ax.set_xlabel('T-SNE Component 1', color='#aaa', fontsize=11)
ax.set_ylabel('T-SNE Component 2', color='#aaa', fontsize=11)
ax.tick_params(colors='#555')
for sp in ax.spines.values(): sp.set_edgecolor('#333355')
ax.legend(facecolor='#0f3460', edgecolor='#333355', labelcolor='white',
          fontsize=10, markerscale=2, loc='best', framealpha=0.3)

plt.tight_layout()
tsne_out = os.path.join(OUTPUTS_DIR, 'tsne_visualization.png')
plt.savefig(tsne_out, dpi=150, bbox_inches='tight')
plt.show()
print(f'✅ Saved → {tsne_out}')

# ── Nhận xét Milestone ────────────────────────────────────────────────────────
print('\n' + '='*55)
print('📌  MILESTONE CHECK — Giai đoạn 1')
print('='*55)
print('→  Nếu các màu tụ lại thành cụm rõ ràng (Apparel / Footwear / Accessories')
print('   không trộn lẫn nhau) → Embeddings ổn → PASS ✅')
print('→  Nếu tất cả trộn lẫn → Kiểm tra lại model hoặc data preprocessing ❌')
print('='*55)
```

## 🔍 Bước 6 — Kiểm tra chất lượng Recommendation (Visual Test)

```python
def show_recommendations(query_idx, top_k=5, filter_category=None):
    """
    Cho một ảnh query, tìm top-k ảnh giống nhất trong FAISS.
    Có thể lọc theo category (ví dụ chỉ lấy 'Footwear' khi query là 'Apparel').
    """
    q_item  = metadata[str(query_idx)]
    q_feat  = embeddings[query_idx:query_idx+1].astype(np.float32)

    # Tìm nhiều hơn để có đủ sau khi lọc
    fetch_k = top_k * 10 if filter_category else top_k + 1
    sims, idxs = index.search(q_feat, fetch_k)

    results = []
    for sim, idx in zip(sims[0], idxs[0]):
        if idx == query_idx: continue  # bỏ chính nó
        item = metadata.get(str(idx))
        if item is None: continue
        if filter_category and item['category'] != filter_category: continue
        results.append((sim, idx, item))
        if len(results) == top_k: break

    # ── Plot ──────────────────────────────────────────────────────────────────
    n_cols = top_k + 1
    fig, axes = plt.subplots(1, n_cols, figsize=(3 * n_cols, 4),
                             facecolor='#1a1a2e')
    fig.suptitle(
        f"Query: {q_item['name'][:40]}\n"
        f"Category: {q_item['category']}  |  Color: {q_item['color']}",
        color='white', fontsize=12, fontweight='bold'
    )

    def show_img(ax, path, title, border='white'):
        try:
            ax.imshow(Image.open(path).convert('RGB'))
        except Exception:
            ax.text(0.5, 0.5, 'No image', ha='center', va='center',
                    transform=ax.transAxes, color='white')
        ax.set_title(title, fontsize=8, color='white', wrap=True)
        ax.axis('off')
        for sp in ax.spines.values():
            sp.set_edgecolor(border); sp.set_linewidth(2)
        ax.set_facecolor('#16213e')

    show_img(axes[0], q_item['local_image_path'],
             f'QUERY\n{q_item["category"]}', border='#F28E2B')

    for i, (sim, idx, item) in enumerate(results):
        show_img(axes[i+1], item['local_image_path'],
                 f'#{i+1}  sim={sim:.2f}\n{item["name"][:20]}\n{item["category"]}',
                 border='#4E79A7')

    plt.tight_layout()
    plt.show()
    return results

# ── Test 1: Tìm ảnh tương tự cùng loại ───────────────────────────────────────
print('🧪 Test 1: Tìm ảnh tương tự trong toàn bộ dataset')
test_idx = random.randint(0, len(df_proc) - 1)
results  = show_recommendations(test_idx, top_k=5)

# ── Test 2: Mix & Match preview — Áo → gợi ý Giày ───────────────────────────
print('\n🧪 Test 2: Mix & Match preview — cho áo vào, lấy giày ra')
apparel_items = [i for i, m in metadata.items() if m['category'] == 'Apparel']
footwear_items = [i for i, m in metadata.items() if m['category'] == 'Footwear']

if apparel_items and footwear_items:
    q = int(random.choice(apparel_items))
    print(f'   Query: {metadata[str(q)]["name"]}  [{metadata[str(q)]["category"]}]')
    _ = show_recommendations(q, top_k=4, filter_category='Footwear')
else:
    print('   Không đủ dữ liệu Apparel/Footwear trong subset này.')
```

```python
# ── Test 3: Color Filter — tìm ảnh màu đỏ trong Footwear ─────────────────────
print('🎨 Test 3: Color filter — Footwear màu Black')

target_color    = 'Black'
target_category = 'Footwear'

matches = [
    (int(k), v) for k, v in metadata.items()
    if v['category'] == target_category
    and target_color.lower() in v['color'].lower()
]
print(f'   Tìm thấy {len(matches)} sản phẩm {target_category} màu {target_color}')

if matches:
    sample = random.sample(matches, min(6, len(matches)))
    fig, axes = plt.subplots(1, len(sample), figsize=(3*len(sample), 4),
                             facecolor='#1a1a2e')
    fig.suptitle(f'{target_category} màu {target_color}', color='white',
                 fontsize=13, fontweight='bold')
    for ax, (idx, item) in zip(axes, sample):
        try:
            ax.imshow(Image.open(item['local_image_path']).convert('RGB'))
        except Exception:
            pass
        ax.set_title(item['name'][:20], fontsize=8, color='white')
        ax.axis('off')
    plt.tight_layout(); plt.show()
```
